## Task 1: Working with Text Files

First, we need some data to work with. We will write a Python script to create a file named `students.txt` containing raw text records.

```python
# Subtask 1.1: Create Sample Text File

# We define the content we want in our file
student_data = """John Smith - Computer Science - Grade: A
Sarah Johnson - Mathematics - Grade: B+
Mike Davis - Physics - Grade: A
Emily Brown - Chemistry - Grade: B
David Wilson - Biology - Grade: A
Lisa Garcia - Engineering - Grade: B+"""

# We open (or create) 'students.txt' in write mode ('w')
with open('students.txt', 'w', encoding='utf-8') as f:
    f.write(student_data)

print("✅ students.txt has been created successfully!")
```

### Subtask 1.2: Read Text File Line by Line
Now we will read that file. Since it's a plain text file, we have to manually "split" the strings to get the specific parts we want (Name, Subject, Grade).

```python
def read_student_file(filename):
    """
    Read student information from a text file line by line.
    """
    try:
        # 'r' stands for read mode
        with open(filename, 'r', encoding='utf-8') as file:
            print("Reading student information:")
            print("-" * 40)

            line_number = 1
            for line in file:
                # .strip() removes whitespace and the newline character (\n)
                clean_line = line.strip()

                if clean_line:
                    print(f"Line {line_number}: {clean_line}")

                    # We split the line by the ' - ' separator
                    parts = clean_line.split(' - ')
                    if len(parts) == 3:
                        name = parts[0]
                        subject = parts[1]
                        # Remove 'Grade: ' to just get the letter
                        grade = parts[2].replace('Grade: ', '')

                        print(f"  Name: {name}")
                        print(f"  Subject: {subject}")
                        print(f"  Grade: {grade}")
                        print()

                    line_number += 1

    except FileNotFoundError:
        print(f"❌ Error: File '{filename}' not found.")
    except Exception as e:
        print(f"❌ Error reading file: {e}")

# Execute the function
read_student_file("students.txt")
```

### Subtask 1.3: Enhanced Text File Processing
Let's take it a step further. We'll convert those letter grades into numbers so we can calculate a class average GPA.

```python
def convert_grade_to_number(grade_str):
    """Convert letter grades to numerical values using a dictionary map."""
    grade_map = {
        'A': 4.0, 'A-': 3.7, 'B+': 3.3, 'B': 3.0, 'B-': 2.7,
        'C+': 2.3, 'C': 2.0, 'C-': 1.7, 'D+': 1.3, 'D': 1.0, 'F': 0.0
    }
    return grade_map.get(grade_str, None)

def process_student_data(filename):
    students = []
    subjects = {}
    grades = []

    try:
        with open(filename, 'r', encoding='utf-8') as file:
            for line in file:
                clean_line = line.strip()
                if clean_line:
                    parts = clean_line.split(' - ')
                    if len(parts) == 3:
                        name = parts[0]
                        subject = parts[1]
                        grade_str = parts[2].replace('Grade: ', '')

                        # Store data in a dictionary for easy access
                        student_info = {'name': name, 'subject': subject, 'grade': grade_str}
                        students.append(student_info)

                        # Count occurrences of each subject
                        subjects[subject] = subjects.get(subject, 0) + 1

                        # Convert and store numerical grades
                        grade_value = convert_grade_to_number(grade_str)
                        if grade_value is not None:
                            grades.append(grade_value)

        # Display Summary Statistics
        print(f"Total students: {len(students)}")
        print("\nSubject distribution:")
        for sub, count in subjects.items():
            print(f"  {sub}: {count} student(s)")

        if grades:
            avg_grade = sum(grades) / len(grades)
            print(f"\nAverage grade (numerical): {avg_grade:.2f}")
            print(f"Highest grade: {max(grades)}")
            print(f"Lowest grade: {min(grades)}")

    except Exception as e:
        print(f"❌ Error: {e}")

process_student_data("students.txt")
```

## Task 2: Working with CSV Files

CSV files are the standard for tabular data. We will create a `sales_data.csv` file and use the `csv` module to read it.

```python
# Subtask 2.1: Create Sample CSV File
import csv

sales_content = [
    ['Date', 'Product', 'Category', 'Quantity', 'Price', 'Sales_Rep'],
    ['2024-01-15', 'Laptop', 'Electronics', '2', '999.99', 'John Smith'],
    ['2024-01-16', 'Mouse', 'Electronics', '5', '25.50', 'Sarah Johnson'],
    ['2024-01-17', 'Keyboard', 'Electronics', '3', '75.00', 'John Smith'],
    ['2024-01-18', 'Monitor', 'Electronics', '1', '299.99', 'Mike Davis'],
    ['2024-01-19', 'Tablet', 'Electronics', '4', '399.99', 'Sarah Johnson'],
    ['2024-01-20', 'Headphones', 'Electronics', '6', '89.99', 'Emily Brown'],
    ['2024-01-21', 'Smartphone', 'Electronics', '2', '699.99', 'John Smith'],
    ['2024-01-22', 'Charger', 'Electronics', '8', '19.99', 'Mike Davis'],
    ['2024-01-23', 'Speaker', 'Electronics', '3', '149.99', 'Sarah Johnson'],
    ['2024-01-24', 'Webcam', 'Electronics', '2', '79.99', 'Emily Brown']
]

with open('sales_data.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(sales_content)

print("✅ sales_data.csv has been created!")
```

### Subtask 2.2: Read CSV File Using csv Module
The `csv.DictReader` tool is very useful because it allows us to access data using the column headers as keys.

```python
import csv

def read_sales_data(filename):
    sales_data = []
    try:
        with open(filename, 'r', newline='', encoding='utf-8') as csvfile:
            # DictReader treats the first row as headers
            csv_reader = csv.DictReader(csvfile)

            print(f"{'Date':<12} {'Product':<12} {'Quantity':<8} {'Price':<10} {'Sales Rep':<15}")
            print("-" * 60)

            for row in csv_reader:
                # Convert strings to numbers for calculations
                quantity = int(row['Quantity'])
                price = float(row['Price'])

                print(f"{row['Date']:<12} {row['Product']:<12} {quantity:<8} ${price:<9.2f} {row['Sales_Rep']:<15}")

                row['Quantity'] = quantity
                row['Price'] = price
                row['Total'] = quantity * price
                sales_data.append(row)

            print("-" * 60)
            print(f"Total records processed: {len(sales_data)}")
        return sales_data
    except Exception as e:
        print(f"❌ Error: {e}")
        return []

sales_list = read_sales_data("sales_data.csv")
```

### Subtask 2.3: Calculate Summary Statistics
Let's calculate which Sales Rep sold the most and which Product is the most popular.

```python
from collections import defaultdict

def calculate_sales_statistics(data):
    if not data:
        return

    total_sales = sum(r['Total'] for r in data)
    sales_by_rep = defaultdict(float)
    sales_by_product = defaultdict(float)

    for r in data:
        sales_by_rep[r['Sales_Rep']] += r['Total']
        sales_by_product[r['Product']] += r['Total']

    print("SALES STATISTICS SUMMARY")
    print("=" * 30)
    print(f"Total Revenue: ${total_sales:,.2f}")
    print("\nSales per Rep:")
    for rep, val in sales_by_rep.items():
        print(f" - {rep}: ${val:,.2f}")

calculate_sales_statistics(sales_list)
```

## Task 3: Write Summary Statistics to New CSV

Generating reports is a key part of file handling. We will now take our calculated data and save it into two new files: a summary report and a detailed report.

```python
import csv
from datetime import datetime

def create_reports(input_file):
    # Process data for Summary
    summary_data = []
    detailed_data = []

    with open(input_file, 'r', newline='', encoding='utf-8') as f:
        reader = list(csv.DictReader(f))

        # Create detailed report data with new columns
        for row in reader:
            qty = int(row['Quantity'])
            prc = float(row['Price'])
            total = qty * prc

            date_obj = datetime.strptime(row['Date'], '%Y-%m-%d')

            detailed_row = row.copy()
            detailed_row['Total_Amount'] = round(total, 2)
            detailed_row['Month'] = date_obj.strftime('%B')
            detailed_row['Day_of_Week'] = date_obj.strftime('%A')
            detailed_data.append(detailed_row)

    # Write Detailed Report
    with open('detailed_sales_report.csv', 'w', newline='', encoding='utf-8') as f:
        fieldnames = ['Date', 'Product', 'Category', 'Quantity', 'Price', 'Total_Amount', 'Sales_Rep', 'Month', 'Day_of_Week']
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(detailed_data)

    # Write Summary Report (Rep level)
    with open('sales_summary.csv', 'w', newline='', encoding='utf-8') as f:
        fieldnames = ['Type', 'Name', 'Total_Sales']
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerow({'Type': 'Global', 'Name': 'All', 'Total_Sales': sum(d['Total_Amount'] for d in detailed_data)})

    print("✅ Reports 'detailed_sales_report.csv' and 'sales_summary.csv' created!")

create_reports('sales_data.csv')
```

## 🛠 Troubleshooting

| Issue | Solution |
| :--- | :--- |
| **FileNotFoundError** | Ensure you ran the code cell that *creates* the file before the one that *reads* it. |
| **Encoding Errors** | Always use `encoding='utf-8'` when opening files to support special characters. |
| **CSV Columns Missing** | Check that your `csv.DictWriter` `fieldnames` match the keys in your dictionary exactly. |
| **Colab Files Disappeared** | Colab storage is temporary. If you restart the runtime, you must run the creation cells again. |
