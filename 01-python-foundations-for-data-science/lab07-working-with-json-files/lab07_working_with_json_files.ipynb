## Task 1: Reading JSON Files and Displaying Formatted Output

### Subtask 1.1: Create Sample JSON Data
We'll start by creating a file named `student_data.json` using Python's file writing capabilities.

```python
# Define the JSON content as a string
json_content = """{
  "student_id": "STU001",
  "name": "Alice Johnson",
  "age": 20,
  "major": "Computer Science",
  "gpa": 3.85,
  "courses": [
    {
      "course_code": "CS101",
      "course_name": "Introduction to Programming",
      "credits": 3,
      "grade": "A"
    },
    {
      "course_code": "MATH201",
      "course_name": "Calculus II",
      "credits": 4,
      "grade": "B+"
    },
    {
      "course_code": "ENG102",
      "course_name": "Technical Writing",
      "credits": 2,
      "grade": "A-"
    }
  ],
  "contact": {
    "email": "alice.johnson@university.edu",
    "phone": "555-0123",
    "address": {
      "street": "123 Campus Drive",
      "city": "University Town",
      "state": "CA",
      "zip": "90210"
    }
  },
  "active": true
}"""

# Write the string to a file
with open('student_data.json', 'w', encoding='utf-8') as f:
    f.write(json_content)

print("✅ student_data.json created successfully!")
```

### Subtask 1.2 & 1.3: Create and Run the JSON Reader
Now we will define a function to read that file and a function to print the information beautifully.

```python
def read_json_file(filename):
    """Reads a JSON file and returns the parsed data safely."""
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            # json.load() converts a file-like object into a Python dictionary
            data = json.load(file)
            return data
    except FileNotFoundError:
        print(f"❌ Error: File '{filename}' not found.")
        return None
    except json.JSONDecodeError as e:
        print(f"❌ Error: Invalid JSON format - {e}")
        return None

def display_student_info(student_data):
    """Prints specific fields from the student dictionary."""
    if not student_data: return

    print("=" * 50)
    print("STUDENT INFORMATION")
    print("=" * 50)
    print(f"Student ID: {student_data['student_id']}")
    print(f"Name: {student_data['name']}")
    print(f"Major: {student_data['major']}")
    print(f"GPA: {student_data['gpa']}")

    # Accessing nested dictionary (contact -> email)
    print(f"Email: {student_data['contact']['email']}")

    print("\nENROLLED COURSES:")
    # Accessing list of dictionaries
    for course in student_data['courses']:
        print(f"- {course['course_code']}: {course['course_name']} ({course['grade']})")

# Main Execution
data = read_json_file("student_data.json")
if data:
    display_student_info(data)
    print("\n--- RAW DATA PREVIEW (Using pprint) ---")
    pprint.pprint(data, indent=2)
```

## Task 2: Converting Dictionary to JSON String and Writing to File

In this task, we go the other way: starting with Python data and saving it as a JSON file.

```python
def create_library_data():
    """Generates a Python dictionary with library information."""
    return {
        "library_name": "Central University Library",
        "location": "Main Campus",
        "established": 1965,
        "total_books": 150000,
        "digital_resources": True,
        "operating_hours": {
            "monday_friday": "8:00 AM - 10:00 PM",
            "saturday": "9:00 AM - 6:00 PM"
        },
        "departments": [
            {"name": "Reference", "floor": 1, "staff_count": 5},
            {"name": "Circulation", "floor": 1, "staff_count": 8}
        ],
        "last_updated": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

# 1. Create the data
lib_data = create_library_data()

# 2. Convert to a formatted JSON string (dumps = 'dump string')
json_string = json.dumps(lib_data, indent=4)
print("--- Generated JSON String ---")
print(json_string[:300] + "...\n")

# 3. Save to a file (dump = 'dump to file')
with open('library_data.json', 'w', encoding='utf-8') as f:
    json.dump(lib_data, f, indent=4)

print("✅ library_data.json has been saved!")
```

## Task 3: Parsing Nested JSON Structures

### Subtask 3.1 & 3.2: Company Data Analysis
We will now create a very complex JSON file and write a script to extract specific insights from it.

```python
company_json = {
  "company": {
    "name": "TechInnovate Solutions",
    "founded": 2010,
    "headquarters": {
      "address": {"city": "San Francisco", "state": "CA"},
      "coordinates": {"latitude": 37.7749, "longitude": -122.4194}
    },
    "departments": {
      "engineering": {
        "head": "Dr. Sarah Chen",
        "budget": 2500000,
        "teams": {
          "backend": {
            "lead": "Alex Thompson",
            "members": [
              {"name": "David Kumar", "role": "Senior Developer", "experience": 7},
              {"name": "Maria Garcia", "role": "DevOps Engineer", "experience": 4}
            ],
            "technologies": ["Python", "Node.js", "PostgreSQL"]
          }
        }
      }
    },
    "financial": {
      "revenue": {"2022": 5200000, "2023": 6800000}
    }
  }
}

with open('company_data.json', 'w', encoding='utf-8') as f:
    json.dump(company_json, f, indent=2)

def analyze_company(filename):
    with open(filename, 'r') as f:
        data = json.load(f)

    co = data['company']
    print(f"🏢 Company: {co['name']}")
    print(f"📍 Location: {co['headquarters']['address']['city']}")

    # Deeply nested access
    eng_dept = co['departments']['engineering']
    print(f"👨‍🔬 Engineering Head: {eng_dept['head']}")

    # Iterate through members
    print("\n--- Backend Team Members ---")
    members = eng_dept['teams']['backend']['members']
    for m in members:
        print(f"- {m['name']} ({m['role']})")

analyze_company('company_data.json')
```

### Subtask 3.4: Interactive JSON Explorer
Run this cell to start an interactive session! You can type names of keys to 'dive' into the data.

```python
class JSONExplorer:
    def __init__(self, filename):
        with open(filename, 'r') as f: self.data = json.load(f)
        self.current_data = self.data
        self.path = ["Root"]

    def run(self):
        print("\n--- 🧭 JSON Explorer ---")
        print("Commands: [key_name], 'back', 'quit'")
        while True:
            print(f"\n📍 Current: {' -> '.join(self.path)}")
            if isinstance(self.current_data, dict):
                print(f"Keys available: {list(self.current_data.keys())}")
            elif isinstance(self.current_data, list):
                print(f"List contains {len(self.current_data)} items. Enter index 0-{len(self.current_data)-1}")
            else:
                print(f"Value: {self.current_data}")

            cmd = input("Enter command: ").strip()
            if cmd == 'quit': break
            elif cmd == 'back':
                self.current_data = self.data
                self.path = ["Root"]
            elif isinstance(self.current_data, dict) and cmd in self.current_data:
                self.path.append(cmd)
                self.current_data = self.current_data[cmd]
            else:
                print("❌ Invalid key or command.")

# To test, uncomment the line below after running the cell:
# explorer = JSONExplorer('student_data.json'); explorer.run()
```

## Verification
Let's ensure all files were created and are readable.

```python
files = ['student_data.json', 'library_data.json', 'company_data.json']
for f in files:
    exists = os.path.exists(f)
    status = "✅ Found" if exists else "❌ Missing"
    print(f"{f}: {status}")
    if exists:
        with open(f, 'r') as j:
            d = json.load(j)
            print(f"   Successfully loaded {type(d)}")
```

## Troubleshooting (Colab Specific)
- **Files disappear?** If you restart your runtime, files in the local folder are deleted. Simply re-run the 'Create' cells above.
- **JSONDecodeError?** This usually means a comma is missing or you used single quotes `'` instead of double quotes `"` in the JSON text.
- **KeyError?** Make sure you are spelling the key exactly as it appears in the JSON (it is case-sensitive!).

## Conclusion & Key Takeaways
- **JSON** is the standard for web data.
- **`json.load()`** reads a file; **`json.loads()`** reads a string.
- **`json.dump()`** writes to a file; **`json.dumps()`** creates a string.
- Nested data requires accessing keys one level at a time (e.g., `data['outer']['inner']`).
