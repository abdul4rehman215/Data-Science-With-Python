## 🛠 Task 1: Understanding Python Variables and Basic Data Types

Variables are like labeled boxes where we store information. Python needs to know what *kind* of information is inside to handle it correctly.

### Subtask 1.1: Creating Your First Python Program
In this cell, we define four variables representing a student's profile. Notice how we don't have to tell Python the type beforehand; it figures it out automatically!

```python
# Lab 1: Python Variables and Data Types Demo
# This code demonstrates basic Python data types

print("=== Python Data Types Demonstration ===")
print()

# 1. Integer (int): Whole numbers without decimals
student_age = 20
print("Student Age:", student_age)
print("Type of student_age:", type(student_age))
print()

# 2. Float (float): Numbers with decimal points
student_gpa = 3.75
print("Student GPA:", student_gpa)
print("Type of student_gpa:", type(student_gpa))
print()

# 3. String (str): Text data surrounded by quotes
student_name = "Alice Johnson"
print("Student Name:", student_name)
print("Type of student_name:", type(student_name))
print()

# 4. Boolean (bool): Logical values representing True or False
is_enrolled = True
print("Is Enrolled:", is_enrolled)
print("Type of is_enrolled:", type(is_enrolled))
print()

print("=== End of Data Types Demo ===")
```

### Subtask 1.2: Exploring Variable Assignment and Naming Rules
Names must start with a letter or underscore, cannot start with a number, and are case-sensitive.

```python
# Variable Naming Rules and Examples
print("=== Variable Naming Examples ===")
print()

# Valid variable names using 'snake_case'
first_name = "John"
last_name = "Smith"
age_in_years = 25
course_code_2024 = "CS101"
_private_variable = "hidden"

print("Valid variable examples:")
print(f"Name: {first_name} {last_name}, Age: {age_in_years}")
print("Course:", course_code_2024)
print()

# Multiple assignment: Assigning values to many variables at once
x, y, z = 10, 20.5, "Hello"
print("Multiple assignment:")
print("x =", x, "| Type:", type(x))
print("y =", y, "| Type:", type(y))
print("z =", z, "| Type:", type(z))
print()

# Same value to multiple variables
a = b = c = 100
print("Same value assignment:")
print("a =", a, "| b =", b, "| c =", c)
```

## 🔄 Task 3: Type Conversion Operations

Sometimes we need to change data from one type to another—like changing the text "20" into the math-ready number 20. This is called **Casting**.

### Subtask 3.1: Basic Type Conversion
Pay close attention to the **Boolean** examples at the bottom of this cell!

```python
# Type Conversion Demonstration
print("=== Type Conversion Examples ===")
print()

# 1. String to Integer: Use int()
age_str = "25"
age_int = int(age_str)
print("Original string:", age_str, "| Converted to int:", age_int)
print()

# 2. String to Float: Use float()
price_str = "19.99"
price_float = float(price_str)
print("Original string:", price_str, "| Converted to float:", price_float)
print()

# 3. Number to String: Use str()
# Useful for displaying numbers inside sentences
score = 95
score_str = str(score)
print("Original integer:", score, "| Converted to string:", score_str)
print()

# 4. Float to Integer: This chops off the decimal (truncation)
gpa = 3.87
gpa_int = int(gpa)
print("Original float:", gpa, "| Converted to int:", gpa_int)
print("Note: Decimal part is truncated, not rounded!")
print()

# 5. Boolean conversions
print("Boolean Conversion Examples:")
print("int(True):", int(True))    # True becomes 1
print("int(False):", int(False))  # False becomes 0
print("bool(1):", bool(1))        # 1 becomes True
print("bool(0):", bool(0))        # 0 becomes False
print()

# CRITICAL RULE: In Python, any non-empty string is True
print("bool('Hello'):", bool("Hello"))
print("bool(''):", bool(""))        # Empty string is False
print("bool('0'):", bool("0"))      # String containing '0' is True!
print("bool('False'):", bool("False")) # String containing 'False' is True!
```

### Subtask 3.2: Interactive Calculator with Type Conversion
This script takes user input, converts it to floats so we can do math, and displays the results.

```python
# Interactive Calculator with Type Conversion
print("=== Simple Calculator ===")
print()

# Step 1: Get user input (comes in as strings)
num1_str = input("Enter the first number (e.g., 15.5): ")
num2_str = input("Enter the second number (e.g., 4.2): ")

# Step 2: Convert strings to floats for calculation
# If we didn't do this, '10' + '5' would equal '105' (text combination)!
num1 = float(num1_str)
num2 = float(num2_str)
print()

# Step 3: Perform calculations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2

print("Calculation Results:")
print(f"{num1} + {num2} = {addition}")
print(f"{num1} - {num2} = {subtraction}")
print(f"{num1} * {num2} = {multiplication}")
print(f"{num1} / {num2} = {division}")
print()

# Step 4: Convert results back to different types for demo
print("Addition as integer (rounded down):", int(addition))
print("Addition as string (for labels):", "The result is " + str(addition))
```

## 🧮 Task 4: Basic Operations and Practical Applications

Now that we know the types, let's see how they interact using math and text operations.

### Subtask 4.1: Arithmetic Operations with Different Data Types
Python can handle complex math and even 'text math' (String operations).

```python
# Basic Operations with Different Data Types
print("=== Arithmetic Operations ===")
print()

# 1. Integer operations
a = 10
b = 3
print("Integer Operations:")
print(f"Addition: {a + b}, Subtraction: {a - b}")
print(f"Division (Standard): {a / b}")
print(f"Floor Division (No remainder): {a // b}")
print(f"Modulus (Remainder only): {a % b}")
print(f"Exponentiation (10 to the power of 3): {a ** b}")
print()

# 2. Float operations
x = 15.5
y = 2.5
print("Float Operations:")
print(f"{x} * {y} = {x * y}")
print()

# 3. Mixed operations (int and float)
# Python 'upgrades' the result to a float if a float is involved
result1 = a + x
print(f"int + float: {a} + {x} = {result1} | Type: {type(result1)}")
print()

# 4. String operations
first_name = "John"
last_name = "Doe"
print("String Operations:")
print("Concatenation:", first_name + " " + last_name) # Adding strings joins them
print("Repetition:", first_name * 3)                # Multiplying strings repeats them
print("Length:", len(first_name))                   # len() counts characters
```

### Subtask 4.2: Practical Application - Student Grade Calculator
This combines everything: input, type conversion, math, and logic.

```python
# Student Grade Calculator
print("=== Student Grade Calculator ===")
print()

# Collect student information
student_name = input("Enter student name: ")
student_id = input("Enter student ID: ")
course_name = input("Enter course name: ")
print()

print("Enter your scores (0-100):")
# Nested conversion: input() inside float()
quiz1 = float(input("Quiz 1 score: "))
quiz2 = float(input("Quiz 2 score: "))
midterm = float(input("Midterm score: "))
final = float(input("Final exam score: "))
project = float(input("Project score: "))

# Calculation logic (Weights: Quizzes 20%, Midterm 25%, Final 35%, Project 20%)
quiz_average = (quiz1 + quiz2) / 2
weighted_score = (quiz_average * 0.20) + (midterm * 0.25) + (final * 0.35) + (project * 0.20)

# Determine letter grade using 'if' statements
if weighted_score >= 90: letter_grade = "A"
elif weighted_score >= 80: letter_grade = "B"
elif weighted_score >= 70: letter_grade = "C"
elif weighted_score >= 60: letter_grade = "D"
else: letter_grade = "F"

# Passing status (Boolean result)
is_passing = weighted_score >= 60

print("\n=== Final Grade Report ===")
print(f"Student: {student_name} (ID: {student_id})")
print(f"Course: {course_name}")
print(f"Weighted Final Score: {weighted_score:.2f}")
print(f"Letter Grade: {letter_grade}")
print(f"Passing Status: {is_passing}")
```

## 🛑 Task 5: Error Handling and Troubleshooting

Programs often break when a user enters something unexpected. We use `try` and `except` to catch these 'crashes' before they happen.

```python
# Common Type Conversion Errors and Solutions
print("=== Error Handling Examples ===")
print()

# Example 1: Safe integer conversion
user_input = input("Enter a whole number (try typing a word to see what happens): ")
try:
    number = int(user_input)
    print(f"Success! {number} is an integer.")
except ValueError:
    # This code runs if the conversion fails
    print(f"Error: '{user_input}' is not a whole number!")
print()

# Example 3: Reminding ourselves of the Boolean Trap
print("Boolean Conversion Behavior Reminder:")
test_values = ["", "0", "False"]
for val in test_values:
    print(f"bool('{val}') result is: {bool(val)}")
```

## 🔍 Troubleshooting Guide
| Issue | Possible Solution |
| :--- | :--- |
| **Cell doesn't run** | Ensure you clicked the Play button or pressed `Shift + Enter`. Check if 'Connecting' is shown at the top right. |
| **IndentationError** | Python is picky! Ensure lines inside `if` or `try` blocks are indented exactly 4 spaces. |
| **ValueError** | You tried to turn text (like "abc") into a number. Use digits only for `int()` and `float()`. |
| **Stuck/Waiting** | If a cell is spinning, it might be waiting for you to type in an input box. Look for a text box below the cell. |
| **ZeroDivisionError** | You can't divide by zero! Make sure your second number in division isn't 0. |
