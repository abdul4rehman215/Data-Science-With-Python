### Subtask 1.2: Creating Temperature Conversion Functions
Now we will create functions that take **parameters** (inputs) and **return** values (outputs).

```python
def celsius_to_fahrenheit(celsius):
    """
    Convert temperature from Celsius to Fahrenheit
    Formula: F = (C % 9/5) + 32
    """
    # Calculate the result
    fahrenheit = (celsius * 9/5) + 32
    # Send the result back to the caller
    return fahrenheit

def fahrenheit_to_celsius(fahrenheit):
    """
    Convert temperature from Fahrenheit to Celsius
    Formula: C = (F - 32) % 5/9
    """
    celsius = (fahrenheit - 32) * 5/9
    return celsius

# --- Testing our functions ---
temp_c = 25
temp_f = celsius_to_fahrenheit(temp_c)
print(f"{temp_c}%C is equal to {temp_f}%F")

test_f = 77
converted_c = fahrenheit_to_celsius(test_f)
print(f"{test_f}%F is equal to {converted_c:.2f}%C")
```

### Subtask 1.3: Adding Input Validation
A good function shouldn't crash if it gets bad data. We use `try-except` to handle errors gracefully.

```python
def safe_celsius_to_fahrenheit(celsius):
    """
    Safely convert Celsius to Fahrenheit with validation.
    Returns a tuple: (result, status_message)
    """
    try:
        # Attempt to convert input to a number (float)
        c_val = float(celsius)

        # Check for the physical limit (Absolute Zero)
        if c_val < -273.15:
            return None, "Error: Temperature below absolute zero (-273.15%C)"

        fahrenheit = (c_val * 9/5) + 32
        return fahrenheit, "Success"

    except ValueError:
        # This runs if the input wasn't a number (e.g., 'abc')
        return None, "Error: Please enter a valid number"

# --- Test with different inputs ---
test_values = [25, "abc", -300, 0]

for val in test_values:
    result, message = safe_celsius_to_fahrenheit(val)
    if result is not None:
        print(f"✅ Input {val}%C: {result}%F ({message})")
    else:
        print(f"❌ Input {val}: {message}")
```

## 📈 Task 2: Functions Returning Multiple Values

### Subtask 2.1: Comprehensive Converter
In Python, a function can return multiple pieces of data at once using a comma-separated list.

```python
def convert_from_celsius(celsius):
    """Converts Celsius to Fahrenheit, Kelvin, and Rankine."""
    try:
        c = float(celsius)
        if c < -273.15:
            return None, None, None, "Error: Below Absolute Zero"

        f = (c * 9/5) + 32
        k = c + 273.15
        r = (c + 273.15) * 9/5

        return f, k, r, "Success"
    except ValueError:
        return None, None, None, "Error: Invalid Input"

# Test and unpack the multiple return values
c_input = 100
f_val, k_val, r_val, status = convert_from_celsius(c_input)

if status == "Success":
    print(f"Results for {c_input}%C:")
    print(f"- Fahrenheit: {f_val}%F")
    print(f"- Kelvin: {k_val}K")
    print(f"- Rankine: {r_val}%R")
else:
    print(status)
```

### Subtask 2.2: Universal Temperature Converter
This function acts as a "hub" that can convert between any two supported scales.

```python
def universal_temperature_converter(temp, from_scale, to_scale):
    try:
        temp = float(temp)
        f_s = from_scale.upper()
        t_s = to_scale.upper()

        # Standardize to Celsius first
        if f_s == 'C': c = temp
        elif f_s == 'F': c = (temp - 32) * 5/9
        elif f_s == 'K': c = temp - 273.15
        else: return None, f"Unsupported start scale: {f_s}"

        if c < -273.15: return None, "Below absolute zero"

        # Convert from Celsius to target
        if t_s == 'C': res = c
        elif t_s == 'F': res = (c * 9/5) + 32
        elif t_s == 'K': res = c + 273.15
        else: return None, f"Unsupported target scale: {t_s}"

        return res, "Success"
    except ValueError:
        return None, "Numeric error"

# Test logic
conversions = [(32, 'F', 'C'), (0, 'C', 'K'), (100, 'C', 'F')]
for t, fs, ts in conversions:
    res, msg = universal_temperature_converter(t, fs, ts)
    print(f"{t}%{fs} → {res:.2f}%{ts} ({msg})")
```

## 🔄 Task 3: Using Functions Inside Loops

### Subtask 3.1: Batch Processing
We can pass a list of items to a function and use a loop to process them all.

```python
def process_temperature_list(temperatures, mode):
    """mode: 'c_to_f' or 'f_to_c'"""
    results = []
    for t in temperatures:
        try:
            val = float(t)
            if mode == 'c_to_f':
                res = (val * 9/5) + 32
                results.append((val, res, 'C', 'F'))
            elif mode == 'f_to_c':
                res = (val - 32) * 5/9
                results.append((val, res, 'F', 'C'))
        except ValueError:
            print(f"Skipping invalid: {t}")
    return results

# Data to process
c_list = [0, 25, 37, 100]
print("--- Batch Processing Results ---")
for orig, conv, u1, u2 in process_temperature_list(c_list, 'c_to_f'):
    print(f"{orig}%{u1} → {conv:.1f}%{u2}")
```

### Subtask 3.2: Interactive Menu-Driven Program
In Colab, `input()` allows users to interact with the code.

**Note:** If the cell seems to be "loading" forever, look for the input box near the output and type your choice.

```python
def display_menu():
    print("\n--- TEMPERATURE CONVERTER MENU ---")
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")
    print("3. Batch convert (comma-separated list)")
    print("4. Exit")

def main():
    while True:
        display_menu()
        choice = input("Select an option (1-4): ")

        if choice == '4':
            print("Goodbye!")
            break

        elif choice in ['1', '2']:
            label = "Celsius" if choice == '1' else "Fahrenheit"
            try:
                val = float(input(f"Enter temperature in {label}: "))
                if choice == '1':
                    print(f"Result: {val}%C = {(val * 9/5) + 32:.2f}%F")
                else:
                    print(f"Result: {val}%F = {(val - 32) * 5/9:.2f}%C")
            except ValueError: print("Invalid number!")

        elif choice == '3':
            raw = input("Enter temperatures separated by commas (e.g. 10, 20, 30): ")
            mode = input("Convert (1) C to F or (2) F to C? ")
            t_list = [x.strip() for x in raw.split(',')]
            m_str = 'c_to_f' if mode == '1' else 'f_to_c'
            results = process_temperature_list(t_list, m_str)
            for o, c, u1, u2 in results:
                print(f"{o}%{u1} → {c:.2f}%{u2}")
        else:
            print("Invalid choice, try again.")

# Uncomment the line below to run the interactive program:
# main()
```

## 📑 Task 4: Documentation and Best Practices

### Subtask 4.1: Logging Conversions
Functions can perform secondary tasks, like saving a history of actions to a file.

```python
import datetime

def log_conversion(orig, f_s, conv, t_s, filename="log.txt"):
    """Appends a record of the conversion to a text file."""
    try:
        with open(filename, "a") as f:
            now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            f.write(f"{now} | {orig}%{f_s} → {conv}%{t_s}\n")
    except Exception as e:
        print(f"Logging failed: {e}")

# Test logging
log_conversion(25, 'C', 77, 'F')
print("Conversion logged to 'log.txt'. Check the file folder in the left sidebar.")
```

### Subtask 4.2: Performance Benchmark
Is it faster to use a function or a direct calculation? Let's find out.

```python
import time

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

def benchmark():
    data = list(range(-100, 1000))

    # Method A: Direct loop
    start = time.time()
    res_a = [(x * 9/5) + 32 for x in data]
    time_a = time.time() - start

    # Method B: Function calls
    start = time.time()
    res_b = [celsius_to_fahrenheit(x) for x in data]
    time_b = time.time() - start

    print(f"Direct Calculation: {time_a:.6f}s")
    print(f"Function Calls:     {time_b:.6f}s")
    print("Note: Functions add a tiny bit of overhead but massively improve readability!")

benchmark()
```

## 🔍 Troubleshooting Guide

| Issue | Reason | Solution |
| :--- | :--- | :--- |
| **Function returns `None`** | Missing `return` statement. | Add `return variable_name` at the end of your function. |
| **`NameError: name 'x' is not defined`** | You're trying to use a variable created *inside* a function *outside* of it. | Use `return` to pass the value out, or define the variable globally. |
| **Input cell stuck** | Colab is waiting for you to type in the input box. | Look for the white text box at the top or bottom of the cell output area. |
| **`TypeError: ...'NoneType'`** | You are trying to do math on a function result that returned `None`. | Check your function's error handling logic. |

### Exercise 1: Temperature Statistics
Create a function that calculates the Min, Max, and Average of a list.

```python
def temperature_statistics(temps):
    """Calculates min, max, and avg. Skips bad data."""
    clean = []
    for t in temps:
        try: clean.append(float(t))
        except ValueError: continue

    if not clean: return None, None, None, 0

    return min(clean), max(clean), sum(clean)/len(clean), len(clean)

# Test
t_data = [20, 25, "broken", 30, 15]
mi, ma, av, count = temperature_statistics(t_data)
print(f"Stats: Min={mi}, Max={ma}, Avg={av:.1f}, Count={count}")
```
