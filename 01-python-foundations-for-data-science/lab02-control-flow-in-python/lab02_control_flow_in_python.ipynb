## 💡 What is Control Flow?
Imagine you are walking to school. If it starts raining, you open your umbrella. If it's sunny, you put on sunglasses. If you see a friend, you stop to say hi. Otherwise, you just keep walking.

**Control Flow** is how a computer program makes those same kinds of decisions. Instead of just running every line of code from top to bottom, the program uses "logic" to decide which parts of the code to run and which parts to skip based on the situation.

### 🚀 Why it Matters
Without control flow, programs would be like a train stuck on a single track—it can only go one way. With control flow, your program becomes a car that can take turns, stop at red lights, or take a detour when it hits traffic. This makes your code smart and interactive!

## 🛠️ Task 1: Understanding Basic Conditional Statements

### Subtask 1.1: Simple If Statement
The `if` statement checks a condition. If the condition is **True**, the indented code runs. If it's **False**, Python skips it.

```python
# Simple if statement example
# We use int() to convert the text input into a whole number
age_input = input("Enter your age: ")
age = int(age_input if age_input else 0) # Fallback to 0 if empty

if age >= 18:
    # This block only runs if age is 18 or older
    print("You are an adult!")
    print("You can vote in elections.")

# This line is NOT indented, so it runs no matter what
print("Program finished.")
```

### Subtask 1.2: If-Else Statement
The `else` statement acts as a "fallback." It runs only if the `if` condition was **False**.

```python
# If-else statement example
score_input = input("Enter your test score (0-100): ")
score = int(score_input if score_input else 0)

if score >= 60:
    # Runs if score is 60 or higher
    print("Congratulations! You passed the test.")
    print(f"Your score: {score}")
else:
    # Runs if score is less than 60
    print("Sorry, you did not pass the test.")
    print(f"Your score: {score}")
    print("Better luck next time!")

print("End of grade report.")
```

### Subtask 1.3: Multiple Conditions with Elif
Use `elif` (short for "else if") when you have more than two possibilities to check.

```python
# Multiple conditions with elif
score_input = input("Enter your test score (0-100): ")
score = int(score_input if score_input else 0)

# Python checks these from top to bottom and stops at the first True one
if score >= 90:
    grade = "A"
    message = "Excellent work!"
elif score >= 80:
    grade = "B"
    message = "Good job!"
elif score >= 70:
    grade = "C"
    message = "Satisfactory performance."
elif score >= 60:
    grade = "D"
    message = "You passed, but consider studying more."
else:
    grade = "F"
    message = "You need to retake the test."

print(f"Score: {score}")
print(f"Grade: {grade}")
print(f"Comment: {message}")
```

## 🧠 Task 2: Working with Logical Operators

### Subtask 2.1: Using the AND Operator
The `and` operator requires **all** conditions to be True.

```python
# Using logical AND operator
username = input("Enter username (try 'student'): ")
password = input("Enter password (try 'python123'): ")
age_val = input("Enter your age: ")
age = int(age_val if age_val else 0)

# All three conditions must be True for access
if username == "student" and password == "python123" and age >= 13:
    print("Access granted!")
    print("Welcome to the Python learning platform.")
else:
    print("Access denied!")

    # Checking specific reasons for failure
    if username != "student":
        print("- Invalid username")

    if password != "python123":
        print("- Invalid password")

    if age < 13:
        print("- You must be at least 13 years old")
```

### Subtask 2.2: Using the OR Operator
The `or` operator requires **at least one** condition to be True.

```python
# Using logical OR operator
day = input("What day is it? ").lower().strip()

if day == "saturday" or day == "sunday":
    print("It's the weekend! Time to relax.")
else:
    print("It's a weekday. Time for school or work.")

# Numeric range example
num_val = input("Enter a number: ")
number = int(num_val if num_val else 0)

if number < 0 or number > 100:
    print("Number is outside the range 0-100")
else:
    print("Number is within the acceptable range")
```

### Subtask 2.3: Using the NOT Operator
The `not` operator **reverses** the True/False value.

```python
# Using logical NOT operator
is_raining = input("Is it raining? (yes/no): ").lower().strip()

# If it is NOT raining, go outside
if not is_raining == "yes":
    print("Great! You can go outside without an umbrella.")
else:
    print("Better take an umbrella!")
```

## 🌡️ Task 3: Temperature Categorization Program

### Subtask 3.1: Basic Temperature Categories
We will use `float()` here because temperatures often have decimals.

```python
# Temperature Categorization Program
print("=== Temperature Categorizer ===")

temp_val = input("Enter temperature in Fahrenheit: ")
temperature = float(temp_val if temp_val else 0.0)

if temperature >= 80:
    category = "hot"
    print(f"{temperature}°F is considered HOT")
elif temperature <= 50:
    category = "cold"
    print(f"{temperature}°F is considered COLD")
else:
    category = "mild"
    print(f"{temperature}°F is considered MILD")

print(f"Temperature category: {category.upper()}")
```

### Subtask 3.2: Enhanced Version (Nested Conditions)
Here we put `if` statements **inside** other `if` statements to handle humidity and sunlight.

```python
print("=== Advanced Temperature Categorizer ===\n")

# Collect Inputs
temp_in = input("Enter temperature in Fahrenheit: ")
temperature = float(temp_in if temp_in else 70.0)

hum_in = input("Enter humidity percentage (0-100): ")
humidity = float(hum_in if hum_in else 50.0)

is_sunny = input("Is it sunny? (yes/no): ").lower().strip()

comfort_level = "Unknown"

print("\n--- Analysis ---")

if temperature >= 80:
    print(f"{temperature}°F is HOT")
    # Nested logic for humidity
    if humidity > 70:
        comfort_level = "Very Uncomfortable (Humid)"
    elif humidity > 50:
        comfort_level = "Uncomfortable"
    else:
        comfort_level = "Tolerable (Dry Heat)"

elif temperature <= 50:
    print(f"{temperature}°F is COLD")
    if temperature <= 32:
        print("Below freezing! Ice may form.")
        comfort_level = "Very Cold"
    else:
        comfort_level = "Chilly"

else:
    print(f"{temperature}°F is MILD")
    if 65 <= temperature <= 75 and humidity < 60 and is_sunny == "yes":
        comfort_level = "Perfect"
    else:
        comfort_level = "Comfortable"

print(f"Comfort Level: {comfort_level}")

# Recommendations using Logical Operators
print("\n--- Recommendations ---")
if temperature > 85 and humidity > 70:
    print("• Stay indoors with AC and drink water.")
elif temperature < 40 or (temperature < 50 and is_sunny == "no"):
    print("• Wear warm clothing; consider staying inside.")
elif 60 <= temperature <= 75 and humidity < 70:
    print("• Great weather for outdoor activities!")
else:
    print("• Dress appropriately for the day.")
```

### Subtask 3.3: Adding Input Validation
Real programs shouldn't crash if a user types a word instead of a number. We use `try-except` to handle errors safely.

```python
# Validated version
def get_valid_temp():
    while True:
        try:
            val = float(input("Enter temperature (-100 to 150): "))
            if -100 <= val <= 150: return val
            print("Please stay within range.")
        except ValueError:
            print("Invalid input! Please enter a number.")

temperature = get_valid_temp()
print(f"Validated Temperature: {temperature}°F")
```

## 🔍 Troubleshooting Guide (Colab Specific)

| Issue | Cause | Solution |
| :--- | :--- | :--- |
| **IndentationError** | Missing 4 spaces after a colon `:` | Ensure code inside `if` blocks is shifted right. |
| **SyntaxError** | Missing `:` at the end of `if` | Check the end of your conditional lines. |
| **Input Stuck** | Waiting for user to press Enter | Look for the text box at the top or under the cell. |
| **NameError** | Variable not defined | Run previous cells where the variable was created. |
