## 📥 Task 2: Loading CSV Files into DataFrames

**ELI10 Explanation:** Imagine a CSV file is just a text version of an Excel sheet. We are going to take that text and turn it into a 'DataFrame'—which is just a fancy name for a table that Python can understand and manipulate.

```python
import pandas as pd
import numpy as np
from io import StringIO

# Subtask 2.1: Create Sample Dataset
# We use StringIO to simulate a physical CSV file directly in our code
students_csv = StringIO("""student_id,name,age,grade,subject,score
1,Alice Johnson,20,A,Math,95
2,Bob Smith,19,B,Math,87
3,Carol Davis,21,A,Science,92
4,David Wilson,20,B,Science,78
5,Emma Brown,19,A,English,88
6,Frank Miller,22,C,English,76
7,Grace Lee,20,B,Math,91
8,Henry Taylor,21,A,Science,85
9,Ivy Chen,19,A,English,94
10,Jack Anderson,20,B,Math,82
""")

# Subtask 2.3: Load CSV File into DataFrame
# read_csv() is the standard way to turn CSV data into a DataFrame
df = pd.read_csv(students_csv)

print("DataFrame loaded successfully!")
display(df)
```

### Subtask 2.4: Understanding DataFrame Structure
**Key Concept:** A DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. Think of it as a spreadsheet or SQL table in Python.

```python
# Check the type of the object to confirm it is a pandas DataFrame
print(f"Type of df: {type(df)}")

# Check the shape (rows, columns) to see the size of our data
print(f"Shape: {df.shape}")

# Check column names to see what variables we have
print(f"Columns: {df.columns.tolist()}")

# Check index to see how rows are labeled
print(f"Index: {df.index}")
```

## 🔍 Task 3: Inspecting Data Using DataFrame Methods

**ELI10 Explanation:** When you get a huge set of data, you don't want to look at all 1,000,000 rows at once. These tools let you peek at the top, the bottom, and get a quick 'health check' of your data.

```python
# Subtask 3.1: Using .head() Method
# .head(n) allows us to preview the first n rows
print("First 5 rows (default):")
display(df.head())

print("\nFirst 3 rows:")
display(df.head(3))

# Subtask 3.2: Using .tail() Method
# .tail(n) allows us to preview the last n rows
print("\nLast 3 rows:")
display(df.tail(3))
```

```python
# Subtask 3.3: Using .info() Method
# This provides a technical summary, showing data types and missing values
print("DataFrame Info:")
df.info()
```

```python
# Subtask 3.4: Using .describe() Method
# This calculates statistics like Mean, Min, and Max for numeric columns
print("Statistical Summary (Numeric Only):")
display(df.describe())

print("\nAll columns summary (Includes names/grades):")
display(df.describe(include='all'))
```

## 🧪 Task 4: Accessing Rows and Columns

**ELI10 Explanation:** This is like using your finger to point at a specific cell in a printed table. We can select just one column, a few specific rows, or even filter for specific people (like everyone with a score over 90).

```python
# Subtask 4.1: Basic Column Access

# Method 1: Using column name as attribute (dot notation)
print("Names column (attribute access):")
print(df.name.head())

# Method 2: Using square brackets (standard way)
print("\nNames column (bracket access):")
print(df['name'].head())

# Method 3: Selecting multiple columns (pass a list of names)
print("\nMultiple columns (name, age, score):")
display(df[['name', 'age', 'score']])
```

```python
# Subtask 4.2: Using .loc[] for Label-Based Selection

# Select row with index label 0
print("Row with index 0:")
print(df.loc[0])

# Select rows 0 to 2 and specific columns
print("\nSpecific rows (0-2) and columns (name, score):")
display(df.loc[0:2, ['name', 'score']])
```

```python
# Subtask 4.3: Using .iloc[] for Position-Based Selection

# Select first 3 rows using integer positions
print("First 3 rows using .iloc:")
display(df.iloc[0:3])

# Select rows 0-2 and columns 1-3 by position
print("\nRows 0-2, columns 1-3:")
display(df.iloc[0:3, 1:4])
```

```python
# Subtask 4.4: Advanced Selection Examples

# Select rows where score > 90 (Filtering)
print("Students with score > 90:")
high_scorers = df.loc[df['score'] > 90]
display(high_scorers)

# Multiple conditions: Age 20 AND Grade A
print("\nStudents aged 20 with grade A:")
result = df.loc[(df['age'] == 20) & (df['grade'] == 'A')]
display(result)
```

## 📊 Task 5: Practical Data Exploration Exercise

**ELI10 Explanation:** Now that you have the tools, let's act like a real Data Analyst! We will create a sales dataset and try to find out which products are making the most money.

```python
# Subtask 5.1: Create a Larger Dataset
sales_csv = StringIO("""product_id,product_name,category,price,quantity_sold,sales_date,region
1,Laptop Pro,Electronics,1299.99,45,2024-01-15,North
2,Wireless Mouse,Electronics,29.99,120,2024-01-15,South
3,Office Chair,Furniture,199.99,67,2024-01-16,East
4,Smartphone X,Electronics,899.99,89,2024-01-16,West
5,Desk Lamp,Furniture,49.99,34,2024-01-17,North
6,Tablet Mini,Electronics,399.99,78,2024-01-17,South
7,Bookshelf,Furniture,149.99,23,2024-01-18,East
8,Headphones,Electronics,79.99,156,2024-01-18,West
9,Standing Desk,Furniture,299.99,12,2024-01-19,North
10,Keyboard Pro,Electronics,129.99,98,2024-01-19,South
11,Monitor 24inch,Electronics,249.99,67,2024-01-20,East
12,Coffee Table,Furniture,179.99,45,2024-01-20,West
13,Webcam HD,Electronics,89.99,87,2024-01-21,North
14,Filing Cabinet,Furniture,119.99,29,2024-01-21,South
15,Printer All-in-One,Electronics,199.99,56,2024-01-22,East
""")

# Subtask 5.2: Load and Explore
sales_df = pd.read_csv(sales_csv)

print("Sales Dataset Loaded!")
display(sales_df.head())
```

```python
# Subtask 5.3: Advanced Data Analysis

# 1. Find products with highest sales quantity
print("Top 5 products by quantity sold:")
top_products = sales_df.loc[:, ['product_name', 'quantity_sold']].sort_values('quantity_sold', ascending=False)
display(top_products.head())

# 2. Filtering by category
print("\nElectronics products:")
electronics = sales_df.loc[sales_df['category'] == 'Electronics']
display(electronics[['product_name', 'price', 'quantity_sold']])
```

```python
# Subtask 5.4: Calculate Revenue
# We create a new column by multiplying price and quantity
sales_df['revenue'] = sales_df['price'] * sales_df['quantity_sold']

print("Products with calculated revenue:")
display(sales_df[['product_name', 'price', 'quantity_sold', 'revenue']].head())

# Category-wise revenue analysis using summing
electronics_revenue = sales_df.loc[sales_df['category'] == 'Electronics', 'revenue'].sum()
furniture_revenue = sales_df.loc[sales_df['category'] == 'Furniture', 'revenue'].sum()

print(f"\nElectronics total revenue: ${electronics_revenue:,.2f}")
print(f"Furniture total revenue: ${furniture_revenue:,.2f}")
```

## 📊 Task 6: Common DataFrame Operations

**ELI10 Explanation:** Sometimes you want to see the most expensive items first (Sorting) or you only want to see items in a specific price range (Filtering).

```python
# Subtask 6.1: Sorting Data
print("Sorted by price (ascending):")
display(sales_df.sort_values('price')[['product_name', 'price']].head())

# Subtask 6.2: Filtering Data with Multiple Conditions
print("\nProducts from East or West regions (using isin):")
regions_filter = sales_df.loc[sales_df['region'].isin(['East', 'West'])]
display(regions_filter[['product_name', 'region']])
```
