## 📥 Creating and Loading Data
Before we can analyze data, we need data! We will create two datasets: one with hospital details and one with their locations.

```python
# Subtask 1.2: Create Sample Hospital Data

# Define the hospital information
hospital_data = {
    'hospital_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'hospital_name': ['City General Hospital', 'St. Mary Medical Center', 'Regional Health System', 'Community Hospital', 'University Medical Center', 'Memorial Hospital', 'Sacred Heart Hospital', 'Metro General'],
    'bed_count': [450, 320, 680, 180, 520, 290, 410, 380],
    'patient_satisfaction': [4.2, 4.7, 3.9, 4.5, 4.1, 4.8, 4.3, 3.8],
    'emergency_services': [True, True, True, False, True, True, True, True],
    'specialty': ['General', 'Cardiac', 'Trauma', 'Pediatric', 'Research', 'Geriatric', 'Cardiac', 'General']
}

# Define the location information mapped by hospital_id
location_data = {
    'hospital_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego'],
    'state': ['NY', 'CA', 'IL', 'TX', 'AZ', 'PA', 'TX', 'CA'],
    'zip_code': ['10001', '90210', '60601', '77001', '85001', '19101', '78201', '92101'],
    'region': ['Northeast', 'West', 'Midwest', 'South', 'West', 'Northeast', 'South', 'West']
}

# Create Pandas DataFrames from dictionaries
hospitals_df = pd.DataFrame(hospital_data)
locations_df = pd.DataFrame(location_data)

# Save these to CSV files so we can practice loading them
hospitals_df.to_csv('hospitals.csv', index=False)
locations_df.to_csv('locations.csv', index=False)

print("Sample CSV files created successfully!")
```

```python
# Subtask 1.3: Load CSV Files

# Reading the files we just created
hospitals = pd.read_csv('hospitals.csv')
locations = pd.read_csv('locations.csv')

print("Hospitals Dataset Loaded:")
display(hospitals.head(3))

print("\nLocations Dataset Loaded:")
display(locations.head(3))
```

## 🔎 Filtering Data
### 💡 ELI10: Filtering
Filtering is like using a sieve. You set a rule (like "only show hospitals with > 400 beds") and Pandas throws away everything that doesn't fit.

```python
# Subtask 2.1 & 2.2: Basic and Multiple Condition Filtering

# 1. Single Condition: Large hospitals
large_hospitals = hospitals[hospitals['bed_count'] > 400]
print("Hospitals with > 400 beds:")
display(large_hospitals[['hospital_name', 'bed_count']])

# 2. Multiple Conditions using '&' (AND)
# We use parentheses () around each condition to avoid errors
large_emergency = hospitals[
    (hospitals['bed_count'] > 300) &
    (hospitals['emergency_services'] == True)
]
print("\nLarge hospitals with emergency services:")
display(large_emergency[['hospital_name', 'bed_count', 'emergency_services']])
```

```python
# Subtask 2.3: String-Based Filtering

# Filtering using string methods
general_hospitals = hospitals[hospitals['hospital_name'].str.contains('General', na=False)]
print("Hospitals with 'General' in the name:")
display(general_hospitals[['hospital_name']])

# Using .isin() for multiple categories
cardiac_trauma = hospitals[hospitals['specialty'].isin(['Cardiac', 'Trauma'])]
print("\nCardiac and Trauma specialty hospitals:")
display(cardiac_trauma[['hospital_name', 'specialty']])
```

## ↕️ Sorting Data
### 💡 ELI10: Sorting
Sorting is just putting things in order, like organizing your bookshelf by height or by the author's name.

```python
# Subtask 3.1 & 3.2: Single and Multiple Column Sorting

# Sort by bed count (highest first)
sorted_by_beds = hospitals.sort_values('bed_count', ascending=False)
print("Hospitals by size (descending):")
display(sorted_by_beds[['hospital_name', 'bed_count']])

# Sort by specialty (A-Z) AND then satisfaction (highest first)
sorted_mixed = hospitals.sort_values(
    ['specialty', 'patient_satisfaction'],
    ascending=[True, False]
)
print("\nSorted by Specialty (Asc) and Satisfaction (Desc):")
display(sorted_mixed[['hospital_name', 'specialty', 'patient_satisfaction']])
```

## 🔗 Merging DataFrames
### 💡 ELI10: Merging (Joins)
Merging is like a puzzle. We use a 'Key' (like `hospital_id`) to connect two different tables together.
*   **Inner**: Only keep rows where the ID is in BOTH tables.
*   **Left**: Keep everything from the first table, add info from the second if it exists.
*   **Outer**: Keep everything from everywhere!

```python
# Task 4: Merging DataFrames

# Inner Join: Combine hospitals and locations based on 'hospital_id'
inner_merged = pd.merge(hospitals, locations, on='hospital_id')
print("Inner Join Result (Hospitals + Locations):")
display(inner_merged[['hospital_name', 'city', 'state', 'region']])

# Merge with Suffixes (Handling overlapping column names)
# Let's create a conflict by adding a 'hospital_name' column to locations
locations_conflict = locations.copy()
locations_conflict['hospital_name'] = locations_conflict['city'] + " Branch"

# Use suffixes to distinguish which 'hospital_name' came from which table
merged_suffixes = pd.merge(
    hospitals,
    locations_conflict,
    on='hospital_id',
    suffixes=('_orig', '_loc_alias')
)
print("\nMerge with Suffixes:")
display(merged_suffixes[['hospital_name_orig', 'hospital_name_loc_alias', 'city']])
```

## 🧪 Full Analysis Pipeline
Now we combine everything: Merge the data, Filter for specific regions, and Sort by satisfaction.

```python
# Step 1: Merge everything
complete_data = pd.merge(hospitals, locations, on='hospital_id')

# Step 2: Filter and Sort
# We want Large hospitals (>300 beds) in West/Northeast with emergency services
final_report = complete_data[
    (complete_data['bed_count'] > 300) &
    (complete_data['emergency_services'] == True) &
    (complete_data['region'].isin(['West', 'Northeast']))
].sort_values('patient_satisfaction', ascending=False)

print("Final Analysis Report:")
display(final_report[['hospital_name', 'city', 'region', 'bed_count', 'patient_satisfaction']])
```

## 📊 Statistical Analysis on Merged Data
### 💡 ELI10: Grouped Stats
Grouping is like sorting your Halloween candy into piles. Once you have a pile for 'Chocolate' and a pile for 'Fruit', you can count which pile is bigger or which has the most pieces.

```python
# Subtask 5.2: Regional Statistics

# Group by region and calculate mean/max bed counts and satisfaction
regional_stats = complete_data.groupby('region').agg({
    'bed_count': ['mean', 'max'],
    'patient_satisfaction': ['mean', 'count']
}).round(2)

# Flatten column names for readability
regional_stats.columns = ['_'.join(col).strip() for col in regional_stats.columns]

print("Regional Hospital Statistics:")
display(regional_stats)

# Find top hospital by patient satisfaction in each region
top_by_region = complete_data.loc[complete_data.groupby('region')['patient_satisfaction'].idxmax()]
print("\nTop rated hospital in each region:")
display(top_by_region[['region', 'hospital_name', 'patient_satisfaction']])
```

## 📋 Creating Summary Reports
### 💡 ELI10: Reporting
A report is a professional summary that answers big questions like "How many high-quality hospitals do we have?"

```python
# Subtask 5.3: Automated Reporting Function

def generate_hospital_report(h_df, l_df):
    # 1. Merge
    merged = pd.merge(h_df, l_df, on='hospital_id')

    # 2. Filter (Satisfaction > 4.0)
    quality_df = merged[merged['patient_satisfaction'] > 4.0]

    # 3. Sort (By size)
    sorted_report = quality_df.sort_values('bed_count', ascending=False)

    # 4. Summary metrics
    metrics = {
        'Total Hospitals': len(merged),
        'High Quality': len(quality_df),
        'Avg Beds': merged['bed_count'].mean()
    }

    return sorted_report, metrics

# Generate and print the report
report, summary = generate_hospital_report(hospitals, locations)
print("--- QUALITY HOSPITAL REPORT ---")
for key, val in summary.items():
    print(f"{key}: {val:.1f}" if isinstance(val, float) else f"{key}: {val}")

display(report[['hospital_name', 'city', 'bed_count', 'patient_satisfaction']])
```
