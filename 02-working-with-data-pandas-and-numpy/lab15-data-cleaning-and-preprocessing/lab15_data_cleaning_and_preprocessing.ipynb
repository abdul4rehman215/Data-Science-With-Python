## 🧹 Cleaning Column Names

**ELI10 Explanation:** Imagine if some of your folder labels were written in messy cursive, some in all caps, and some had extra spaces at the end. It would be hard to find anything! We are going to rename our columns so they all look neat and the same (lowercase with underscores).

```python
# Task 2: Column Renaming to Standard Format

# Analyze current names for issues like spaces or mixed casing
print("Current column names:")
for i, col in enumerate(df.columns):
    print(f"{i+1}. '{col}'")

# Define a mapping dictionary to standardize the names
# Standard: lowercase, no spaces (use underscores instead), remove leading/trailing spaces
column_mapping = {
    'Customer ID': 'customer_id',
    'Customer Name': 'customer_name',
    'Product Category': 'product_category',
    'Product Name': 'product_name',
    'Order Date': 'order_date',
    'Ship Date': 'ship_date',
    'Sales Amount': 'sales_amount',
    'Quantity Ordered': 'quantity_ordered',
    'Unit Price': 'unit_price',
    'Discount %': 'discount_percent',
    'Profit Margin': 'profit_margin',
    'Region ': 'region',
    ' Country': 'country',
    'Unused Column': 'unused_column',
    'Constant Value': 'constant_value'
}

# Apply the renaming. We create a new DataFrame 'df_cleaned' to preserve the original if needed.
df_cleaned = df.rename(columns=column_mapping)

print("\nColumns renamed successfully!")
print("New column names:", df_cleaned.columns.tolist())
```

## 🕒 Fixing Data Types

**ELI10 Explanation:** Computers are picky. If you give a computer a date but tell it it's just a 'sentence' (string), it can't calculate how many days have passed. We need to tell the computer exactly what is a number, what is a date, and what is a category.

```python
# Task 3: Data Type Conversion

# 3.2 Convert date columns from string to datetime objects
# 'errors=coerce' turns invalid dates (like 'invalid_date') into NaT (Not a Time)
print("Converting date columns...")
df_cleaned['order_date'] = pd.to_datetime(df_cleaned['order_date'], format='%Y-%m-%d', errors='coerce')
df_cleaned['ship_date'] = pd.to_datetime(df_cleaned['ship_date'], format='%Y-%m-%d', errors='coerce')

# 3.3 Convert numeric columns
# 'errors=coerce' handles empty strings or text found in numeric columns by making them NaN
print("Converting numeric columns...")
df_cleaned['sales_amount'] = pd.to_numeric(df_cleaned['sales_amount'], errors='coerce')
df_cleaned['unit_price'] = pd.to_numeric(df_cleaned['unit_price'], errors='coerce')
df_cleaned['discount_percent'] = pd.to_numeric(df_cleaned['discount_percent'], errors='coerce')
df_cleaned['profit_margin'] = pd.to_numeric(df_cleaned['profit_margin'], errors='coerce')
# 'Int64' (capital I) allows for whole numbers even if there are missing values (NaN)
df_cleaned['quantity_ordered'] = pd.to_numeric(df_cleaned['quantity_ordered'], errors='coerce').astype('Int64')

# 3.4 Convert categorical columns
# This saves memory and tells pandas these are labels, not just random text
categorical_columns = ['customer_name', 'product_category', 'product_name', 'region', 'country']
for col in categorical_columns:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].astype('category')

print("Data type conversion complete. Updated dtypes:")
print(df_cleaned.dtypes)
```

## 🗑 Removing Bad Data

**ELI10 Explanation:** Now we throw away the trash! If a column has no information at all, or a row is completely empty, it doesn't help us. We'll find those empty spots and remove them.

```python
# Task 4: Removing Unnecessary Columns and Rows

# 4.2 Remove columns that are useless (all null or only one constant value)
columns_to_remove = []
for col in df_cleaned.columns:
    # Identify columns where every single value is missing
    if df_cleaned[col].isnull().all():
        columns_to_remove.append(col)
    # Identify columns where every row has the exact same value (no information gain)
    elif df_cleaned[col].nunique(dropna=True) <= 1:
        columns_to_remove.append(col)

print(f"Removing columns: {columns_to_remove}")
df_cleaned = df_cleaned.drop(columns=columns_to_remove)

# 4.4 Remove rows with only NaN values (entirely empty rows)
initial_rows = len(df_cleaned)
df_cleaned = df_cleaned.dropna(how='all')

# Optional: Remove rows with excessive missing data (e.g., more than 50% of columns are empty)
max_nulls = int(len(df_cleaned.columns) * 0.5)
df_cleaned = df_cleaned[df_cleaned.isnull().sum(axis=1) <= max_nulls]

print(f"Removed {initial_rows - len(df_cleaned)} empty or highly incomplete rows.")
print(f"New shape: {df_cleaned.shape}")
```

## ✅ Validation

**ELI10 Explanation:** This is the 'double-check' phase. We look at our cleaned data to make sure it's actually correct and that we didn't accidentally break anything.

```python
# Task 5: Data Quality Validation

# 5.1 Check if data types match our expectations
print("--- Data Type Validation ---")
display(df_cleaned.dtypes)

# 5.2 Generate a Data Quality Report
print("\n--- Data Quality Report ---")
for col in df_cleaned.columns:
    null_pct = (df_cleaned[col].isnull().sum() / len(df_cleaned)) * 100
    print(f"{col}: {df_cleaned[col].isnull().sum()} missing values ({null_pct:.1f}%)")

# 6.1 Handle Duplicate Records
# Exact duplicates can skew analysis results
duplicates = df_cleaned.duplicated().sum()
if duplicates > 0:
    print(f"\nRemoving {duplicates} duplicate records...")
    df_cleaned = df_cleaned.drop_duplicates()

# 6.2 Data Consistency Checks
# Logical check: Shipping shouldn't happen BEFORE ordering
if 'order_date' in df_cleaned.columns and 'ship_date' in df_cleaned.columns:
    bad_dates = df_cleaned[df_cleaned['ship_date'] < df_cleaned['order_date']]
    if not bad_dates.empty:
        print(f"\n⚠ WARNING: Found {len(bad_dates)} records where ship_date is before order_date!")

# Save the final result
df_cleaned.to_csv('cleaned_sales_data.csv', index=False)
print("\nFinal cleaned dataset saved as 'cleaned_sales_data.csv'")
```

## 🛠 Troubleshooting

**Issue 1: Date Conversion Errors**
If dates are in multiple formats (e.g., 2024-01-01 and 01/01/24), use `pd.to_datetime(df['col'], dayfirst=True)` or a custom function with multiple format attempts.

**Issue 2: Memory Optimization**
For huge datasets, convert `float64` to `float32` and `int64` to `int32` where possible using `df['col'].astype(np.float32)`.

**Issue 3: Missing Column Errors**
Always use `print(df.columns.tolist())` before a task to verify the column names exist and are spelled correctly.

## 📚 Key Takeaways

*   **Standardization is Key:** Consistently named columns make code easier to write and read.
*   **Types Matter:** Mathematical operations only work on numeric types; time-series analysis requires datetime objects.
*   **Garbage In, Garbage Out:** Analysis is only as good as the data. Cleaning prevents wrong conclusions.
*   **Iterative Process:** You might find more issues after your first round of cleaning; always validate your results.
